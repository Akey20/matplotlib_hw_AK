```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"


# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

###mouse_metadata # just a checkpoint to see data
##study_results # just a checkpoint to see data
# Combine the data into a single dataset
mouse_df = pd.merge(mouse_metadata, study_results, how ="outer", on= "Mouse ID")

# Display the data table for preview
mouse_df.head(25)
##mouse_df.shape # used to get number of rows to compare when dropping duplicate below 
```

```python
# Checking the number of mice
##mouse_df["Mouse ID"].value_counts() #checkpoint remove for prod
##mouse_df.shape # checkpoint to see data for compare once remove duplicate below
mice_number = mouse_df["Mouse ID"].nunique()  ## comment out line 4&5 to run shape
mice_number
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_ID = mouse_df.loc[mouse_df.duplicated(subset=["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
duplicate_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
##duplicate_mouse_df # mouse_df.loc[mouse_df["Mouse ID"] == "g989", :]
##duplicate_mouse_df  # either will work 2&3 or lines 4&5 but going with 4&5 for production
mouse_index_df = mouse_df.set_index("Mouse ID")
mouse_index_df.loc["g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_df = mouse_df[mouse_df["Mouse ID"].isin(duplicate_ID)==False]
clean_mouse_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse_number = clean_mouse_df["Mouse ID"].nunique()
clean_mouse_number
##clean_mouse_df.shape # this is a check to compare to the last shape count after dup removed
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_mean = clean_mouse_df["Tumor Volume (mm3)"].groupby(clean_mouse_df["Drug Regimen"]).mean()
##tumor_mean #check- rmove for production
tumor_median = clean_mouse_df["Tumor Volume (mm3)"].groupby(clean_mouse_df["Drug Regimen"]).median()
##tumor_median #check- rmove for production
tumor_variance = clean_mouse_df["Tumor Volume (mm3)"].groupby(clean_mouse_df["Drug Regimen"]).var()
##tumor_variance #check- rmove for production
tumor_std = clean_mouse_df["Tumor Volume (mm3)"].groupby(clean_mouse_df["Drug Regimen"]).std()
##tumor_std #check- rmove for production
tumor_sem = clean_mouse_df["Tumor Volume (mm3)"].groupby(clean_mouse_df["Drug Regimen"]).sem()
##tumor_sem #check- rmove for production

# Assemble the resulting series into a single summary dataframe.
tumor_mouse_df = pd.DataFrame({"Tumor Mean": tumor_mean, "Tumor Median": tumor_median,
                              "Tumor Variance": tumor_variance, "Tumor Standard Dev": tumor_std,
                              "Tumor SEM": tumor_sem})
tumor_mouse_df

```

```python
#- Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
aggregation_summary = clean_mouse_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].agg(["mean", "median",
                                                                                     "var", "std", "sem"])
aggregation_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_mouse_df.groupby(["Mouse ID"])["Timepoint"].max()
#setup dataframe
last_timepoint_df = pd.DataFrame(last_timepoint)

last_timepoint_df
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
#merge on mouse id and timepoint -inner 
last_timepointfinal_df = pd.merge(clean_mouse_df, last_timepoint_df, on=("Mouse ID", "Timepoint"), how = 'inner')
last_timepointfinal_df
```

```python
# Put treatments into a list for for loop (and later for plot labels) list per readme should only have 4 drugs Capomulin,
#Ramicane, Infubinol, Ceftamin
drugs =["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
drugscomb = mouse_df[mouse_df["Drug Regimen"].isin(drugs)]



# Create empty list to fill with tumor vol data (for plotting)
tumor_capo=[]
tumor_rami=[]
tumor_infu=[]
tumor_ceft=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
for index, row in last_timepointfinal_df.iterrows():
    if row["Drug Regimen"] == "Capomulin":
        tumor_capo.append(row["Tumor Volume (mm3)"])
    if row["Drug Regimen"] == "Ramicane":
        tumor_rami.append(row["Tumor Volume (mm3)"])    
    if row["Drug Regimen"] == "Infubinol":
        tumor_infu.append(row["Tumor Volume (mm3)"])
    if row["Drug Regimen"] == "Ceftamin":
        tumor_ceft.append(row["Tumor Volume (mm3)"])    
    
    
    # add subset 
drugs_df = pd.DataFrame({drugs[0]: tumor_capo,
                        drugs[1]: tumor_rami,
                        drugs[2]: tumor_infu,
                        drugs[3]: tumor_ceft})    
    
    
    # Determine outliers using upper and lower bounds
drugs_df_upper =drugs_df.max()   
drugs_df_lower =drugs_df.min()  
drugs_df
    
```

```python
#Determine outliers
#calculations for quartiles, quantiles
quartile_capomulin = drugs_df[drugs[0]].quantile([.25, .5, .75])
lower_capomulin = quartile_capomulin[.25]
med_capomulin = quartile_capomulin[.5]
upper_capomulin = quartile_capomulin[.75]
iqr_capomulin = upper_capomulin - lower_capomulin
lowerq_capomulin = lower_capomulin -(1.5*iqr_capomulin)
upperq_capomulin = upper_capomulin + (1.5*iqr_capomulin)
max_capomulin = drugs_df[drugs[0]].max()
min_capomulin = drugs_df[drugs[0]].min()

quartile_ramicane = drugs_df[drugs[1]].quantile([.25, .5, .75])
lower_ramicane = quartile_ramicane[.25]
med_ramicane = quartile_ramicane[.5]
upper_ramicane = quartile_ramicane[.75]
iqr_ramicane = upper_ramicane - lower_ramicane
lowerq_ramicane = lower_ramicane -(1.5*iqr_ramicane)
upperq_ramicane = upper_ramicane + (1.5*iqr_ramicane)
max_ramicane = drugs_df[drugs[0]].max()
min_ramicane = drugs_df[drugs[0]].min()

quartile_infubinol = drugs_df[drugs[2]].quantile([.25, .5, .75])
lower_infubinol= quartile_infubinol[.25]
med_infubinol = quartile_infubinol[.5]
upper_infubinol = quartile_infubinol[.75]
iqr_infubinol = upper_infubinol - lower_infubinol
lowerq_infubinol = lower_infubinol -(1.5*iqr_infubinol)
upperq_infubinol = upper_infubinol+ (1.5*iqr_infubinol)
max_infubinol = drugs_df[drugs[0]].max()
min_infubinol = drugs_df[drugs[0]].min()

quartile_ceftamin = drugs_df[drugs[3]].quantile([.25, .5, .75])
lower_ceftamin= quartile_ceftamin[.25]
med_ceftamin = quartile_ceftamin[.5]
upper_ceftamin = quartile_ceftamin[.75]
iqr_ceftamin = upper_ceftamin - lower_ceftamin
lowerq_ceftamin = lower_ceftamin -(1.5*iqr_ceftamin)
upperq_ceftamin = upper_ceftamin+ (1.5*iqr_ceftamin)
max_ceftamin = drugs_df[drugs[0]].max()
min_ceftamin = drugs_df[drugs[0]].min()

outliers = pd.DataFrame({"Treatments": drugs, "Lower Quartile": [lower_capomulin, lower_ramicane, lower_infubinol, lower_ceftamin],
                         "Upper Quartile": [upper_capomulin, upper_ramicane, upper_infubinol, upper_ceftamin],
                         "IQR": [iqr_capomulin, iqr_ramicane,iqr_infubinol,iqr_ceftamin],
                         "Median": [med_capomulin, med_ramicane, med_infubinol, med_ceftamin],
                         "Upper Bounds": [upperq_capomulin, upperq_ramicane, upperq_infubinol, upperq_ceftamin],
                         "Lower Bounds": [lowerq_capomulin, lowerq_ramicane, lowerq_infubinol, lowerq_ceftamin],
                         "Max": [max_capomulin, max_ramicane, max_infubinol, max_ceftamin],
                         "Min": [min_capomulin, min_ramicane, min_infubinol, min_ceftamin]})
outliers
outliers_group = outliers.groupby('Treatments')
outliers_group.sum()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen


# only use the Capomulin regimen
capo_df = clean_mouse_df.loc[clean_mouse_df['Drug Regimen'] == 'Capomulin']

# Find average tumor volume for each mouse

avg_vol_df = pd.DataFrame(capo_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().sort_values()).reset_index().rename(columns={'Tumor Volume (mm3)': 'avg_tumor_vol'})

# Merge average tumor volume onto data_df and drop duplicates
avg_vol_df = pd.merge(capo_df, avg_vol_df, on='Mouse ID')
final_avg_vol_df = avg_vol_df[['Weight (g)', 'avg_tumor_vol']].drop_duplicates()
final_avg_vol_df

x = final_avg_vol_df['Weight (g)']
y = final_avg_vol_df['avg_tumor_vol']

# Create a scatter plot based on new dataframe above with circle markers and listed colors
plt.scatter(x, y)

# Add labels and title to plot
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Average Tumor Volume by Weight')
# Display plot
plt.show()

avg_vol_df
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regim
#use same data code from above
correlation = st.pearsonr(x,y)
#print answer from st.pearson formula above
print(f"""The correlation between weight and the avg tumor volume (mm3) on the Capolulin regimen is {round(correlation
       [0],2)}""")

#linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(x,y)
regress_values = x * slope + intercept
line_eq = "y =" + str(round(slope,2)) + "x + " + str(round(intercept,2))

#plot data, format and add labels
plt.scatter(x,y)
plt.plot(x,regress_values, "m--")
plt.annotate(line_eq,(20,37),fontsize=15,color="purple")
plt.xlabel("Weight(g)")
plt.ylabel("Avg Tumor Volume(mm3)")
plt.title("Average Tumor Volume by Weight")

print(f"The r-squared is: {rvalue**2}")

plt.show()

```
